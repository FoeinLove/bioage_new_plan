```python
import pandas as pd

d1 = pd.read_csv('biomarker.csv')
d2 = pd.read_csv('mast.csv')
d3 = pd.read_csv('surveys.csv')

d2 = d2[['Idind','GENDER']]
d3 = d3[(d3.wave == 2009)&(d3.age >=20)&(d3.age<=80)][['Idind','age']]
d4 = d2.merge(d3,on='Idind')
d4.rename({'Idind': 'IDind'}, axis=1, inplace=True)
df = d4.merge(d1,on='IDind')
df .rename({'IDind': 'ID','age':'AGE'}, axis=1, inplace=True)
```

```python
df = df[['ID', 'GENDER', 'AGE', 'UREA', 'UA', 'APO_A', 'LP_A', 'HS_CRP', 'CRE',
       'HDL_C', 'LDL_C', 'APO_B', 'MG', 'FET', 'INS', 'HGB', 'WBC', 'RBC',
       'PLT','HbA1c', 'TP', 'ALB', 'GLUCOSE', 'TG', 'TC', 'ALT', 
       'TRF', 'TRF_R', 'Y46_1DL', 'wave']]
```
